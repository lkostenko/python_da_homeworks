```python
import datetime
```

```python

def get_time(date):
  correct_date = None
  if len(date) == 19:
    correct_date = datetime.datetime.strptime(date, "%m.%d.%Y %H:%M:%S")
  elif len(date) == 16:
    correct_date = datetime.datetime.strptime(date, "%m.%d.%Y %H:%M")
  return correct_date

```

```python
date_string = "11.30.2022 10:26:37"
date_str = "01.09.2023 11:38"
print(len(date_string), len(date_str))

```

```python
print(get_time(date_str))
```

```python
df['Timestamp'] = df.Timestamp.apply(get_time)
```

12. Зробіть зводну таблицю, де стовпці - це 'Ваш тайтл', рядки - 'Знання англійської мови', а значення - медіанна ЗП. Які висновки можна зробити?
Тут ще для зручності можна зробити градієнтне забарвлення і відформатувати значення для легшого перегляду.

```python
pivot_salary_englevel_title = pd.pivot_table(df, index='Ваш тайтл', columns='Знання англійської мови', values='Зарплата у $$$ за місяць, лише ставка після сплати податків', aggfunc='median')
pivot_salary_englevel_title = pivot_salary_englevel_title.style.format('${:.0f}')
pivot_salary_englevel_title = pivot_salary_englevel_title.background_gradient(axis=None, cmap='YlOrRd')

pivot_salary_englevel_title
```
